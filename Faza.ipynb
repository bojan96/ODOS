```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

Posmatraćemo uticaj početnih faza na oblik signala. Signal sa kojim ćemo eksperimentisati je dat jednačinom

$$\tilde{x} \left( n  \right) = \sum_{i=0}^{3} \frac{1}{2i + 1} \sin \left(\frac{2 \pi}{64} \left(2i+1 \right) n + \phi_i \right).$$

Ovaj signal se sastoji od prva četiri harmonika razvoja povorke pravougaonih impulsa u Furijeov red.

Za početak neka su početne faze svih harmonika jednake nuli.

```python
phi0 = 0
phi1 = 0
phi2 = 0
phi3 = 0
n = np.arange(64)
x = np.sin(2*np.pi/64 * n + phi0) + (1/3)*np.sin(2*np.pi/64 * 3 * n + phi1) + \
    (1/5)*np.sin(2*np.pi/64 * 5 * n + phi2) + (1/7)*np.sin(2*np.pi/64 * 7 * n + phi3)
```

```python
plt.figure()
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x(n)')
```

Pretpostavimo sada da su početne faze harmonika oblika

$$\phi_i = -5 \frac{2 \pi}{64} \left(2i + 1 \right), \; i=0,1,2,3.$$

Ovakve početne faze odgovaraju prolasku signala kroz svepropusni filtar sa linearnom fazom i kašnjenjem od 5 odmjeraka. Oblik signala je nepromijenjen.

```python
phi0 = -2*np.pi/64 * 5
phi1 = -2*np.pi/64 * 3 * 5
phi2 = -2*np.pi/64 * 5 * 5
phi3 = -2*np.pi/64 * 7 * 5
n = np.arange(64)
x = np.sin(2*np.pi/64 * n + phi0) + (1/3)*np.sin(2*np.pi/64 * 3 * n + phi1) + \
    (1/5)*np.sin(2*np.pi/64 * 5 * n + phi2) + (1/7)*np.sin(2*np.pi/64 * 7 * n + phi3)
```

```python
plt.figure()
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x(n)')
```

U slučaju da su početne faze harmonika oblika

$$\phi_i = -\frac{2 i \pi}{3}$$

situacija odgovara prolasku signala kroz sistem sa nelinearnom fazom. Talasni oblik signala je promijenjen iako je njegov amplitudni spektar isti. Sistem je unio linearna izobličenja signala.

```python
phi0 = 0
phi1 = -2*np.pi*1/3
phi2 = -2*np.pi*2/64
phi3 = -2*np.pi*3/64
n = np.arange(64)
x = np.sin(2*np.pi/64 * n + phi0) + (1/3)*np.sin(2*np.pi/64 * 3 * n + phi1) + \
    (1/5)*np.sin(2*np.pi/64 * 5 * n + phi2) + (1/7)*np.sin(2*np.pi/64 * 7 * n + phi3)
```

```python
plt.figure()
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x(n)')
```
