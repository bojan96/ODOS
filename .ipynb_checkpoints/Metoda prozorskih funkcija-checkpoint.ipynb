```python
%matplotlib notebook
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
h = [0, 0, 1, 1, 1, 0, 0]
w = [-1, -0.5, -0.5, 0, 0.5, 0.5, 1]

plt.rc('text', usetex=True)
plt.figure()
plt.plot(w, h)
plt.xticks([-1, -0.5, 0, 0.5, 1])
```

```python
wc = 0.5 # normalizovana sa pi
n = np.arange(-40, 41)
h = wc*np.sinc(0.5*n)

plt.figure()
plt.stem(n, h)
plt.xlabel('n')
plt.ylabel('h(n)')
```

```python
nn = np.arange(-25, 26)
h = 0.5*np.sinc(0.5*nn)
hh = np.concatenate((np.zeros(15), h, np.zeros(15)))

plt.figure()
plt.stem(n, hh)
plt.xlabel('n')
plt.rc('text', usetex=True)
plt.xlabel('$n$')
plt.ylabel(r'$\hat{h}(n)$')
```

```python
nn = np.arange(-25, 26)
h = 0.5*np.sinc(0.5*nn)
n = np.arange(51)

plt.figure()
plt.stem(n, h)
plt.xlabel('n')
plt.rc('text', usetex=True)
plt.xlabel('$n$')
plt.ylabel(r'$\hat{h}(n)$')
```

```python
w, H = signal.freqz(h, 1)

plt.rc('text', usetex=True)
plt.figure()
plt.subplot(211)
plt.plot(w/np.pi, np.abs(H))
plt.ylabel(r'\vert H(e^{j\omega}) \vert')
plt.title('Amplitudna karakteristika')
plt.subplot(212)
plt.plot(w/np.pi, np.angle(H))
plt.xlabel(r'\omega/\pi')
plt.ylabel(r'\phi(\omega)')
plt.title('Fazna karakteristika')
```

```python
N = 4
n = np.arange(-N, N+1)
h = 0.5*np.sinc(0.5*n)
w, H = signal.freqz(h, 1, worN=2048)

plt.rc('text', usetex=True)
plt.figure()
line1, = plt.plot(w/np.pi, np.abs(H))
plt.ylabel(r'\vert H(e^{j\omega}) \vert')
plt.title('Amplitudna karakteristika')
plt.xlabel(r'\omega/\pi')
```

```python
N = 10
n = np.arange(-N, N+1)
h = 0.5*np.sinc(0.5*n)
w, H = signal.freqz(h, 1, worN=2048)

line2, = plt.plot(w/np.pi, np.abs(H))
```

```python
N = 50
n = np.arange(-N, N+1)
h = 0.5*np.sinc(0.5*n)
w, H = signal.freqz(h, 1, worN=2048)

line3, = plt.plot(w/np.pi, np.abs(H))
plt.legend((line1, line2, line3), (r'$N=4$', r'$N=10$', r'$N=50$'))
```

```python
w = [0, 0.5, 0.5, 1]
H = [1, 1, 0, 0]

line4, = plt.plot(w, H, 'r')
plt.legend((line1, line2, line3, line4), (r'$N=4$', r'$N=10$', r'$N=50$', 'idealan'))
```

```python
from scipy.special import diric

N = 10

omega = np.linspace(-np.pi, np.pi, 1001)
xdtft = (2*N+1)*diric(omega, 2*N+1)

plt.rc('text', usetex=True)
plt.figure()
plt.plot(omega/np.pi, xdtft)
plt.xlabel(r'$\omega/\pi$')
plt.ylabel(r'W(e^{j\omega})')
```

```python
from scipy.signal import windows
win_rect = windows.boxcar(2*N+1)
n = np.arange(-N, N+1)
w, win_rect_spec = signal.freqz(win_rect, 1)

plt.figure()
plt.subplot(211)
plt.stem(n, win_rect)
plt.title('Pravougaona prozorska funkcija')
plt.xlabel('n')
plt.ylabel('w(n)')
plt.subplot(212)
plt.plot(w/np.pi, np.abs(win_rect_spec))
plt.xlabel(r'\omega/\pi')
plt.ylabel(r'\vert W(e^{j\omega}) \vert')
```

```python
win_bart = windows.bartlett(2*N+1)
n = np.arange(-N, N+1)
w, win_bart_spec = signal.freqz(win_bart, 1)

plt.figure()
plt.subplot(211)
plt.stem(n, win_bart)
plt.title('Bartletova prozorska funkcija')
plt.xlabel('n')
plt.ylabel('w(n)')
plt.subplot(212)
plt.plot(w/np.pi, np.abs(win_bart_spec))
plt.xlabel(r'\omega/\pi')
plt.ylabel(r'\vert W(e^{j\omega}) \vert')
```

```python
N = 10
win_rect = windows.boxcar(2*N+1)
n = np.arange(-N, N+1)
w, win_rect_spec = signal.freqz(win_rect, 1)
win_rect_spec /= np.max(win_rect_spec)

plt.figure()
line1, = plt.plot(w/np.pi, 20*np.log10(np.abs(win_rect_spec)))
plt.xlabel(r'\omega/\pi')
```

```python
N = 50
win_rect = windows.boxcar(2*N+1)
n = np.arange(-N, N+1)
w, win_rect_spec = signal.freqz(win_rect, 1)
win_rect_spec /= np.max(win_rect_spec)

line2, = plt.plot(w/np.pi, 20*np.log10(np.abs(win_rect_spec)))
plt.legend((line1, line2), ('N=10', 'N=50'))
```

```python
N = 50
win_rect = windows.boxcar(2*N+1)
n = np.arange(-N, N+1)
w, win_rect_spec = signal.freqz(win_rect, 1)
win_rect_spec /= np.max(win_rect_spec)

plt.figure()
line1, = plt.plot(w/np.pi, 20*np.log10(np.abs(win_rect_spec)))
plt.xlabel('\omega/\pi')

win_bart = windows.bartlett(2*N+1)
w, win_bart_spec = signal.freqz(win_bart, 1)
win_bart_spec /= np.max(win_bart_spec)

line2, = plt.plot(w/np.pi, 20*np.log10(np.abs(win_bart_spec)))

win_hamm = windows.hamming(2*N+1)
w, win_hamm_spec = signal.freqz(win_hamm, 1)
win_hamm_spec /= np.max(win_hamm_spec)

line3, = plt.plot(w/np.pi, 20*np.log10(np.abs(win_hamm_spec)))
plt.legend((line1, line2, line3), ('Pravougaoni', 'Bartletov', 'Hemingov'))
```

```python
N = 50
n = np.arange(-N, N+1)
hr = signal.firwin(2*N+1, 0.5, window='boxcar')
hb = signal.firwin(2*N+1, 0.5, window='bartlett')
hh = signal.firwin(2*N+1, 0.5, window='hamming')

w, Hr = signal.freqz(hr, 1)
_, Hb = signal.freqz(hb, 1)
_, Hh = signal.freqz(hh, 1)

plt.figure()
line1, = plt.plot(w/np.pi, 20*np.log10(np.abs(Hr)))
line2, = plt.plot(w/np.pi, 20*np.log10(np.abs(Hb)))
line3, = plt.plot(w/np.pi, 20*np.log10(np.abs(Hh)))
plt.xlabel(r'\omega/\pi')
plt.ylabel(r'\vert H(e^{j \omega}) \vert')
plt.legend((line1, line2, line3), ('Pravougaoni', 'Trougaoni', 'Hemingov'))
```
