```python

%matplotlib notebook
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as pyplot
import numpy.random as random
import sounddevice as sd
```

```python
# 2 Tacka

toneFreq = 440
FS = 48000
N = round((FS - toneFreq / 2) / toneFreq)
print("N: ", N)
b = np.zeros(N + 1)
b[N] = 2
print("b : ", b)
a = np.zeros(N + 2)
(a[0], a[N], a[N + 1]) = (2, -1, -1)
print("a : ", a)
(w, H) = signal.freqz(b, a, 512, True)
pyplot.figure()
pyplot.xlabel("f[Hz]")
pyplot.ylabel("|H(w)|")
freqAxis = (w * FS) / (2 * np.pi)
pyplot.plot(freqAxis, np.abs(H))
pyplot.figure()
pyplot.xlabel("f[Hz]")
pyplot.ylabel("arg H(w)")
pyplot.plot(freqAxis, np.angle(H))
```

```python
# Tacka 3
def karplusStrong(f, fs, l):
    N = round((fs - f / 2) / f)
    (b, a) = (np.zeros(N + 1), np.zeros(N + 2))
    (b[N], a[0], a[N], a[N + 1]) = (2, 2, -1, -1)
    noise = random.normal(0, 1, l * fs)
    result = signal.lfilter(b, a, noise)
    normalized = result / np.amax(np.abs(result))
    return normalized
```

```python
result = karplusStrong(440, 48000, 5)
```
