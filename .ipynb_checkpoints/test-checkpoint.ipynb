```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as mpt

interval = 51
ref = interval // 2
x = np.arange(-ref, ref + (0 if interval % 2 == 0 else 1), 1)
f = 1
Fs = 10 * f
w = 2 * np.pi * f
y = np.cos((w/Fs) * x )
print(y[ref + 1])

mpt.stem(x, y)
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as mpt
import scipy.io.wavfile as wavfile
import sounddevice as sd
import numpy as np

sampleRate, data = wavfile.read("handel44100.wav")
print("Sample rate: ", sampleRate)
print("Sample data type: ", data.dtype)
print("Data shape: ", data.shape)
print("Duration", data.shape[0] / sampleRate)
print("Memory usage: ", data.nbytes)

x = np.linspace(0, 11, data.shape[0])
ch1 = data[:,0]
ch2 = data[:, 1]
mpt.figure()
mpt.subplots_adjust(hspace = 1)
mpt.subplot(2, 1, 1)
mpt.plot(x, ch1)
mpt.title("First channel")
mpt.xlabel("Time")
mpt.subplot(2, 1, 2)
mpt.plot(x, ch2)
mpt.title("Second channel")
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as pyplot
import sounddevice as sd

def sinus(A, F_0, phi, t_start, t_end, F_s):
    t = np.linspace(t_start, t_end ,(t_end - t_start) * F_s)
    s = A * np.sin(2 * np.pi * F_0 * t + phi)
    return (s, t)

F_s = 8000
A = 0.1
F_0 = 400
length = 10
phi = 45 * np.pi / 180
s, t = sinus(A, F_0, phi, 0, length, F_s)

pyplot.figure()
pyplot.subplot(4, 1, 1)
pyplot.plot(t, s)
pyplot.subplot(4, 1, 2)
pyplot.plot(s)
pyplot.subplot(4, 1, 3)
pyplot.stem(t, s)
pyplot.subplot(4, 1, 4)
pyplot.stem(s)

```
